```python
# import required libraries
import pandas as pd
import numpy as np
import pickle
```

```python
# import all the important data
database = pd.read_pickle('databasefrom1.pickle')
dictionaryofkeywords = pd.read_pickle('dictionaryofkeywordsfrom2.pickle')
paragraphs = pd.read_pickle('databasetextcomponentfrom1.pickle')
```

```python
# importance currently is just the count, heading counts as 2, paratraph as 1
```

```python
importancesofkeywords = {}

for heading in database['preprocessed_heading']:
    for word in heading:
        if word not in importancesofkeywords:
            importancesofkeywords[word] = 2
        else:
            importancesofkeywords[word] += 2

for textids in database['paragraphs']:
    for paragraphid in textids:
        for word in paragraphs.iloc[paragraphid]['preprocessed_words']: # more preprocessing is needed in paragraphs
            if word in importancesofkeywords:
                importancesofkeywords[word] += 1
```

```python
# exporting the importances as importancesofkeywords.pickle
with open('importancesofkeywordsfrom3.pickle', 'wb') as file:
    pickle.dump(importancesofkeywords, file)
```
