```python
# function to make 2-grams
def make_2_grams(listofthings):
    lst = []
    for i in range(len(listofthings)-1):
        lst.append([listofthings[i], listofthings[i+1]])
    return lst
```

```python
# function to count 2-grams and return a dictionary
def count_two_grams(headings=None, texts=None, everyheadingcount=1, everytextcount=1):
    twodict = {}
    if headings is not None:
        for heading in headings:
            lst = make_2_grams(heading)
            for twogram in lst:
                if ' '.join(twogram) in twodict:
                    twodict[' '.join(twogram)]+=everyheadingcount
                else:
                    twodict[' '.join(twogram)]=everyheadingcount
    if texts is not None:
        for paragraph in texts:
            for line in paragraph:
                lst = make_2_grams(line)
                for twogram in lst:
                    if ' '.join(twogram) in twodict:
                        twodict[' '.join(twogram)]+=everytextcount
                    else:
                        twodict[' '.join(twogram)]=everytextcount
    return twodict
```

```python
def calculate_kw_importanaces(headings, texts, everyheadingcount=1, everytextcount=1):
    importances = {}
    for heading in headings:
        for kw in heading:
            if kw in importances:
                importances[kw]+=everyheadingcount
            else:
                importances[kw]=everyheadingcount
    for paragraph in texts:
        for line in paragraph:
            for kw in line:
                if kw in importances:
                    importances[kw]+=everytextcount
    return importances
```

```python
# functions for second iteration
def update_heading_second(heading, twodict, importances, certaintyinpercent):
    newheading = []
    i=0
    while i<len(heading):
        one = heading[i]
        try:
            two = heading[i+1]
            candidates = [one, one+' '+two, two]
            oneratio = twodict[candidates[1]]/importances[one]
            tworatio = twodict[candidates[1]]/importances[two]
            if oneratio>certaintyinpercent or tworatio>certaintyinpercent:
                newheading.append(candidates[1])
                i+=2
                continue
            else:
                newheading.apend(one)
                i+=1
                continue
        except:
            newheading.append(one)
            i+=1
            continue
    return newheading

def update_text_second(paragraph, twodict, importances, certainityinpercent=0.40):
    newparagraph = []
    for line in paragraph:
        newparagraph.append(update_heading_second(line, twodict, importances, certainityinpercent))
    return newparagraph

def kwstokps_iteration2(headings, twodict, importances, certainityinpercent=0.40):
    if len(headings)==0:
        return ('nothing to work with.')
    i=0
    while i<len(headings):
#         temptwodict = update_twodict(twodict, heading)
        headings[i] = update_heading_second(headings[i], twodict, importances, certainityinpercent)
        i+=1
    return headings
```
