```python
# import required libraries
import pandas as pd
import numpy as np
import pickle
import networkx as nx
```

```python
# reading the required .pickle files
database = pd.read_pickle('databasefrom4.pickle')
text = pd.read_pickle('databasetextcomponentfrom4.pickle')
G = pd.read_pickle('directedgraphoftopicsfrom5.pickle')
importances = pd.read_pickle('importancesofkeywordsfrom4.pickle')
```

```python
# 1. when there is text under the heading
for headingindex in database.index:
    heading = database.iloc[headingindex]['updated_headings2']
    if len(heading)==0:
        continue
    textids = database.iloc[headingindex]['paragraphs']
    if len(textids)!=0:
        KWms = heading[0]
        for i in range(len(heading)-1):
            if importances[heading[i]]<importances[KWms]:
                KWms = heading[i]
        for textid in textids:
            temptext = text.iloc[textid]['updated_words2']
            for word in temptext:
                if word in importances:
                    if word != KWms:
                        if (word, KWms) in G.edges():
                            G.edges[(word, KWms)]['weight']+=0.5
                        else:
                            G.add_weighted_edges_from([(word, KWms, 0.5)])
```

```python
# 2. when there are n-subheadings under a heading
def call_related_headings(headingid, database, listofrelatedids=None, nextheading=False):
    if listofrelatedids is None:
        listofrelatedids = []
    if not nextheading:
        if len(database.iloc[headingid]['subheadings'])!=0:
            subheadingid = database.iloc[headingid]['subheadings'][0]
            listofrelatedids.append(subheadingid)
            listofrelatedids = call_related_headings(subheadingid, database, listofrelatedids, True)
    if nextheading:
        if len(database.iloc[headingid]['nextheading'])!=0:
            nextheadingid = database.iloc[headingid]['nextheading'][0]
            listofrelatedids.append(nextheadingid)
            listofrelatedids = call_related_headings(nextheadingid, database, listofrelatedids, True)
    return listofrelatedids

for headingindex in database.index:
    heading = database.iloc[headingindex]['updated_headings2']
    if len(heading)==0:
        continue
    subheadingids = call_related_headings(headingindex, database)
    if len(subheadingids)!=0:
        KWms = heading[0]
        for i in range(len(heading)-1):
            if importances[heading[i]]<importances[KWms]:
                KWms = heading[i]
        for subheadingid in subheadingids:
            subheading = database.iloc[subheadingid]['updated_headings2']
            if len(subheading)!=0:
                KWls = subheading[0]
            else:
                continue
            for j in range(len(subheading)-1):
                if importances[subheading[j]]>importances[KWls]:
                    KWls = subheading[j]
            if KWms==KWls:
                pass # write code here later
            if (KWms, KWls) in G.edges():
                G.edges[(KWms, KWls)]['weight'] += 0.75
            else:
                G.add_weighted_edges_from([(KWms, KWls, 0.75)])
```

```python
# deleting bi-directional edges like shown above, keeping the one that has KW1-->KW2 such that importance of KW1
# is more than KW2
trackofedges = {}
### to ensure that the same edge isn't counted twice
for edge in G.edges():
    trackofedges[edge] = False
edgestodelete = []
for node in G.nodes:
    edges = G.edges(node)
    for a, b in edges:
        if not trackofedges[(a, b)]:
            trackofedges[(a, b)] = True
            if (b, a) in edges:
                trackofedges[(b, a)] = True
                weight1 = G.edges[(a, b)]['weight']
                weight2 = G.edges[(b, a)]['weight']
                if importances[a]>=importances[b]: ### there should be something more solid when both the importances
                                        ### are equal, because
                                        ### this affects the final results
                    G.edges[(a, b)]['weight']+=weight2
                    edgestodelete.append((b, a))
                else:
                    G.edges[(b, a)]['weight']+=weight1
                    edgestodelete.append((a, b))
```

```python
G.remove_edges_from(edgestodelete)
```
