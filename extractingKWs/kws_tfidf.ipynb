```python
import pandas as pd
import pickle
from nltk import everygrams
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from IPython.display import clear_output
import seaborn as sns
import matplotlib.pylab as plt
stopwords = pd.read_pickle('/Users/chinmayshrivastava/Desktop/OSCA/V4.0/stopwords.pickle')
```

```python
hvector = CountVectorizer(analyzer='word', ngram_range=(1, 3))
h = hvector.fit_transform(headings.iloc[5000:15000]['headingtext'])
vocab = hvector.vocabulary_
del hvector
del h
```

```python
len(vocab)
```

```python
docsfortfidf = []
for hid in headings.index:
    if hid<5000:
        continue
    temp = ''
    for pgid in headings.iloc[hid]['paragraphs']:
        temp += (' '+paragraphs.iloc[pgid]['text'])
    docsfortfidf.append(temp)
    if hid==14999:
        break
```

```python
pvector = TfidfVectorizer(vocabulary=vocab, ngram_range=(1, 3))
p = pvector.fit_transform(docsfortfidf)
tfcount = p.toarray()
del pvector
del p
del docsfortfidf
```

```python
kws = {}
kwid = 0
hid = 0
```

```python
while True:
    print(hid)
    if len(headings.iloc[hid]['headingtext'])==0:
        hid+=1
        continue
    if len(headings.iloc[hid]['paragraphs'])==0:
        hid+=1
        continue
    try:
        heading = headings.iloc[hid]['headingtext']
        hvector = CountVectorizer(analyzer='word', ngram_range=(1, 3))
        h = hvector.fit_transform([heading])
        headingv = hvector.vocabulary_
    except ValueError:
        hid+=1
        continue
    for element in headingv:
        test = element.split(' ')
        if test[0] in list(stopwords['Word']):
            if test[len(test)-1] in list(stopwords['Word']):
                continue
        if tfcount[hid-5000, vocab[element]]>0.11:
            if element not in kws:
                kws[element] = kwid
                kwid += 1
    if hid==9999:
        break
    hid += 1
    clear_output(wait=True)
```
