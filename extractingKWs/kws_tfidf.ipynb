```python
kws = {}
kwid = 0
hid = 0
```

```python
lower = 0
upper = 5000
while True:
    try:
        hvector = CountVectorizer(analyzer='word', ngram_range=(1, 3))
        h = hvector.fit_transform(headings.iloc[lower:upper]['headingtext'])
        vocab = hvector.vocabulary_
        del hvector
        del h

        docsfortfidf = []
        for hid in range(lower, upper):
            temp = ''
            for pgid in headings.iloc[hid]['paragraphs']:
                temp += (' '+paragraphs.iloc[pgid]['text'])
            docsfortfidf.append(temp)
        pvector = TfidfVectorizer(vocabulary=vocab, ngram_range=(1, 3))
        p = pvector.fit_transform(docsfortfidf)
        tfcount = p.toarray()
        del pvector
        del p
        del docsfortfidf

        for hid in range(lower, upper):
            print(hid)
            if len(headings.iloc[hid]['headingtext'])==0:
                continue
            if len(headings.iloc[hid]['paragraphs'])==0:
                continue
            try:
                heading = headings.iloc[hid]['headingtext']
                hvector = CountVectorizer(analyzer='word', ngram_range=(1, 3))
                h = hvector.fit_transform([heading])
                headingv = hvector.vocabulary_
            except ValueError:
                continue
            for element in headingv:
                test = element.split(' ')
                if test[0] in list(stopwords['Word']):
                    if test[len(test)-1] in list(stopwords['Word']):
                        continue
                if tfcount[hid-lower, vocab[element]]>0.11:
                    if element not in kws:
                        kws[element] = {'id':kwid, 'count':1}
                        kwid += 1
                    else:
                        kws[element]['count'] += 1
                        kwid += 1
        #     if hid==4999:
        #         break
        #     hid += 1
            clear_output(wait=True)

        lower+=5000
        upper+=5000
        if upper>len(headings.index) and lower<len(headings.index):
            upper = len(headings.index)
    except:
        break
```
