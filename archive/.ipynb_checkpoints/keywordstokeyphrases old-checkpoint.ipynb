```python
# searching for all the sub-connected heading components to form keyphrases
def call_related_headings(headingid, database, listofrelatedids=None, nextheading=False):
    if listofrelatedids is None:
        listofrelatedids = []
    if len(database.iloc[headingid]['subheadings'])!=0:
        subheadingid = database.iloc[headingid]['subheadings'][0]
        listofrelatedids.append(subheadingid)
        listofrelatedids = call_related_headings(subheadingid, database, listofrelatedids, True)
    if nextheading:
        if len(database.iloc[headingid]['nextheading'])!=0:
            nextheadingid = database.iloc[headingid]['nextheading'][0]
            listofrelatedids.append(nextheadingid)
            listofrelatedids = call_related_headings(nextheadingid, database, listofrelatedids, True)
    return listofrelatedids

# searching for all the sub-connected text components to form keyphrases
def call_related_text(headingid, database, listofrelatedids=None, nextheading=False):
    if listofrelatedids is None:
        listofrelatedids = []
    for paragraphid in database.iloc[headingid]['paragraphs']:
        listofrelatedids.append(paragraphid)
    if len(database.iloc[headingid]['subheadings'])!=0:
        subheadingid = database.iloc[headingid]['subheadings'][0]
        listofrelatedids = call_related_text(subheadingid, database, listofrelatedids, nextheading=True)
    if nextheading:
        if len(database.iloc[headingid]['nextheading'])!=0:
            nextheadingid = database.iloc[headingid]['nextheading'][0]
            listofrelatedids = call_related_text(nextheadingid, database, listofrelatedids, nextheading=True)
    return listofrelatedids
```

```python
# for each heading call all the associated sub-headings and text
# make one and two grams of the headings
# if these occur in the subheadings/text count them seperately
# in the end restructure the heading into combinations of keywords into keyphrases which makes the most sense 
## according to the ngram count

newheadingdatabase = []
for headingid in database.index:
    headingdata = database.iloc[headingid]['preprocessed_heading']
    if len(headingdata)!=0:
        relatedheadings = call_related_headings(headingid, database)
        relatedtext = call_related_text(headingid, database)
        candidates = set()
        for i in range(len(headingdata)):
            if i!=len(headingdata)-1:
                candidates.add(headingdata[i])
                candidates.add(headingdata[i]+' '+headingdata[i+1])
            else:
                candidates.add(headingdata[i])
        
        temporaryimportances = {}
        for kp in candidates:
            temporaryimportances[kp] = 0
            
        for subheadid in relatedheadings:
            subheaddata = database.iloc[subheadid]['preprocessed_heading']
            i = 0
            while True:
                if i<len(subheaddata)-1:
                    if subheaddata[i]+' '+subheaddata[i+1] in temporaryimportances:
                        temporaryimportances[subheaddata[i]+' '+subheaddata[i+1]]+=1
                        i+=2
                    elif subheaddata[i] in temporaryimportances:
                        temporaryimportances[subheaddata[i]]+=1
                        i+=1
                    else:
                        i+=1
                elif i==len(subheaddata)-1:
                    if subheaddata[i] in temporaryimportances:
                        temporaryimportances[subheaddata[i]]+=1
                    break
                else:
                    break
        
        for textid in relatedtext:
            lines = text.iloc[textid]['preprocessed_lines']
            for line in lines:
                i = 0
                while True:
                    if i<len(line)-1:
                        if line[i]+' '+line[i+1] in temporaryimportances:
                            temporaryimportances[line[i]+' '+line[i+1]]+=0.5
                            i+=2
                        elif line[i] in temporaryimportances:
                            temporaryimportances[line[i]]+=0.5
                            i+=1
                        else:
                            i+=1
                    elif i==len(line)-1:
                        if line[i] in temporaryimportances:
                            temporaryimportances[line[i]]+=0.5
                        break
                    else:
                        break
        
        # updating the database
        newheadingdata = []
        i = 0
        while True:
            if i<len(headingdata)-2:
                one = headingdata[i]
                two = headingdata[i+1]
                three = headingdata[i]+' '+headingdata[i+1]
                four = headingdata[i+1]+' '+headingdata[i+2]
                if temporaryimportances[three]!=0:
                    if temporaryimportances[four]!=0:
                        if temporaryimportances[four]>temporaryimportances[three]:
                            newheadingdata.append(one)
                            newheadingdata.append(four)
                            i+=3
                        else:
                            newheadingdata.append(three)
                            i+=2
                    else:
                        newheadingdata.append(three)
                        i+=2
                else:
                    newheadingdata.append(one)
                    i+=1
            elif i==len(headingdata)-2:
                one = headingdata[i]
                two = headingdata[i+1]
                three = headingdata[i]+' '+headingdata[i+1]
                if temporaryimportances[three]!=0:
                    newheadingdata.append(three)
                    i+=2
                else:
                    newheadingdata.append(one)
                    newheadingdata.append(two)
                    break
            elif i==len(headingdata)-1:
                newheadingdata.append(headingdata[i])
                break
            else:
                break
        
        newheadingdatabase.append(newheadingdata)
    elif len(headingdata)==0:
        newheadingdatabase.append([])

database['updated_headings'] = newheadingdatabase        

### we can iterate here itself
```

```python
# updating all the paragraphs using the kw dictionary, because the last cell only updated each heading individually
newtextdatabase = []
newworddatabase = []
for index in text.index:
    textdata = list(text.iloc[index]['preprocessed_lines'])
    newtextdata = []
    
    for line in textdata:
        newlinedata = []
        i = 0
        while True:
            if i<len(line)-2:
                one = line[i]
                two = line[i+1]
                three = line[i]+' '+line[i+1]
                four = line[i+1]+' '+line[i+2]
                if three in temporarykeyworddictionary:
                    if four in temporaryimportances:
                        if temporarykeyworddictionary[four]>temporarykeyworddictionary[three]:
                            newlinedata.append(one)
                            newlinedata.append(four)
                            i+=3
                        else:
                            newlinedata.append(three)
                            i+=2
                    else:
                        newlinedata.append(three)
                        i+=2
                else:
                    newlinedata.append(one)
                    i+=1
            elif i==len(line)-2:
                one = line[i]
                two = line[i+1]
                three = line[i]+' '+line[i+1]
                if three in temporarykeyworddictionary:
                    newlinedata.append(three)
                    i+=2
                else:
                    newlinedata.append(one)
                    newlinedata.append(two)
                    break
            elif i==len(line)-1:
                newlinedata.append(line[i])
                break
            else:
                break

        newtextdata.append(newlinedata)
    newtextdatabase.append(newtextdata)
    
    newwordset = []
    for element in newtextdata:
        for word in element:
            newwordset.append(word)
    newworddatabase.append(newwordset)
    
text['updated_lines'] = newtextdatabase
text['updated_words'] = newworddatabase
```

```python
# recalculating the importances of keywords
importancesofkeywords = {}

for heading in database['updated_headings']:
    for word in heading:
        if word not in importancesofkeywords:
            importancesofkeywords[word] = 2
        else:
            importancesofkeywords[word] += 2

for textids in database['paragraphs']:
    for paragraphid in textids:
        for word in text.iloc[paragraphid]['updated_words']:
            if word in importancesofkeywords:
                importancesofkeywords[word] += 1
```
