```python
hvector = CountVectorizer(vocabulary=kwunique, ngram_range=(1, 4))
pvector = CountVectorizer(vocabulary=kwunique, ngram_range=(1, 4))
h = hvector.fit_transform(headings['headingtext'])
p = pvector.fit_transform(paragraphs['text'])
hcount = h.toarray()
pcount = p.toarray()
```

```python
def updatedict(somedict):
    tempd = {}
    somedictlist = sorted(somedict, key=len)
    for element in somedictlist:
        temp = everygrams(element.split(' '), max_len=len(element.split(' '))-1)
        temp = sorted(temp, key=len)
        temp = [' '.join(element) for element in temp]
        for ele in temp:
            if ele not in tempd:
                tempd[ele] = somedict[element]
            else:
                continue
    for element in tempd:
        if element in somedict:
            somedict[element] -= tempd[element]
    return somedict

headingcountfactor = 5
paragraphcountfactor = 1

for hid in headings.index:
    clear_output(wait=True)
    print(hid)
    htemp = {}
    ptemp = {}
    for kw1 in kwunique:
        if hcount[hid, kwunique[kw1]]!=0:
            htemp[kw1] = hcount[hid, kwunique[kw1]]
    for pgid in headings.iloc[hid]['paragraphs']:
        for kw2 in kwunique:
            if pcount[pgid, kwunique[kw2]]!=0:
                if kw2 not in ptemp:
                    ptemp[kw2] = pcount[pgid, kwunique[kw2]]
                else:
                    ptemp[kw2] += pcount[pgid, kwunique[kw2]]
    htemp = updatedict(htemp)
    ptemp = updatedict(ptemp)
    sumpara = 0
    for element in ptemp:
        sumpara += ptemp[element]
    for kw1 in htemp:
        kwdata[kwid[kw1]]['synonyms'][kw1] += htemp[kw1]*headingcountfactor
        for kw2 in ptemp:
            kwdata[kwid[kw2]]['synonyms'][kw2] += ptemp[kw1]*paragraphcountfactor
            matrix[kwid[kw1], kwid[kw2]] += ptemp[kw2]/sumpara
```

```python
def getminprereqs(searchid, matrix, discriminant=0.2):
    output = []
    cloud = [[searchid, matrix[searchid, searchid]]]
    track = []
    track.append(searchid)
    while True:
        id_ = cloud.pop(0)[0]
        list_ = list(matrix[id_])
        for element in range(len(list_)):
            if element==id_:
                continue
            if list_[element]>discriminant*max(list_):
                if element not in track:
                    cloud.append([element, list_[element]])
                    track.append(element)
        cloud = sorted(cloud, key=out0, reverse=True)
        output.append(id_)
        if len(cloud)==0:
            break
    return output

def printprereqs(kwdata, searchid, matrix):
    get = getminprereqs(searchid, matrix)
    while True:
        if len(get)!=0:
            print(kwdata[get.pop()]['name'])
        else:
            break
```

```python
# print the n-most related topics
def topnprereqs(searchid, matrix):
    temp = []
    for i in range(len(matrix[searchid])):
        if i!=searchid:
            temp.append([i, matrix[searchid, i]])
    temp = sorted(temp, key=out1, reverse=True)
    return temp
```

```python
import matplotlib.pyplot as plt
   
name_ = 'white paper'
XY = topnprereqs(kwid[name_], matrix)
X = [element for element in range(len(XY))]
Y = [element[1] for element in XY]

plt.plot(X, Y)
plt.title(f"Occurance of '{name_}'")
plt.xlabel('kws')
plt.ylabel('frequency')
plt.show()
# some
```

```python
# 80-20% principle
def gettopconnections(searchid, matrix, kwdata):
    totalsum = sum(sum(matrix))
    returnthis = []
    sigma = sum(matrix[searchid])
    temp = topnprereqs(searchid, matrix)
    tempsum = 0
    i = 0
    while True:
        returnthis.append(temp[i][0])
        tempsum+=temp[i][1]
        P = sum(matrix[:, temp[i][0]])/totalsum
        p = matrix[searchid, temp[i][0]]/sum(matrix[searchid])
        print(f"P({kwdata[temp[i][0]]['name']}):{P}, P({kwdata[temp[i][0]]['name']}|{kwdata[searchid]['name']}):{p}")
        if tempsum>0.2*sigma:
            break
        i+=1
    return returnthis
```
