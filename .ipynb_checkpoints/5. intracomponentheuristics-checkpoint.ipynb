```python
# importing the required libraries
import pandas as pd
import numpy as np
import pickle
import networkx as nx
```

```python
# import the required pickle files
database = pd.read_pickle('databasefrom4.pickle')
text = pd.read_pickle('databasetextcomponentfrom4.pickle')
importances = pd.read_pickle('importancesofkeywordsfrom4.pickle')
```

```python
database.head(60)
```

```python
# initiate a new graph
G = nx.DiGraph()
G
```

```python
# add all keywords as nodes with associated importances
for kw in importances.keys():
    G.add_node(kw, importance=importances[kw])
```

```python
len(G.nodes())
```

```python
# calue for each intra-heading kw-to-kw weight
intraheadingwt = 1
```

```python
# intra-component headings
for heading in database['updated_headings2']:
    if len(heading)>1:
        for i in range(len(heading)-1):
            one = heading[i]
            two = heading[i+1]
            if importances[one]>importances[two]:
                if (one, two) in G.edges():
                    G.edges[(one, two)]['weight']+=intraheadingwt
                else:
                    G.add_weighted_edges_from([(one, two, intraheadingwt)])
            else:
                if (two, one) in G.edges():
                    G.edges[(two, one)]['weight']+=intraheadingwt
                else:
                    G.add_weighted_edges_from([(two, one, intraheadingwt)])
```

```python
len(G.nodes())
```

```python
for edge in G.edges():
    print(edge, G.edges[edge]['weight'])
```

```python
# saving the graph for further analysis
with open('directedgraphoftopicsfrom5.pickle', 'wb') as file:
    pickle.dump(G, file)
```
