```python
# import the required libraries
import pandas as pd
import numpy as np
import pickle
from datacleaningfuncs import *
```

```python
# import database of structured information from books and the web
database = pd.read_pickle('databasefrom1.pickle')
# data containing the document id's and the weights associated with each of them
# documents = pd.read_pickle('documents.pickle')
```

```python
# list of keyword-defining components
listkdc = list(database['preprocessed_heading'])
```

```python
database.head()
```

```python
listkdc = remove_empty_KWs_and_KPs(listkdc)

## experiment - accuracy of lemmatized vs non-lemmatized words
```

```python
# initialising dictionary of keywords
dictionaryofkeywords = {}
```

```python
# findingkeywords
for heading in listkdc:
    for word in heading:
        if word not in dictionaryofkeywords:
            dictionaryofkeywords[word] = True
```

```python
# exporting the keyword dictionary as dictionaryofkeywords.pickle
with open('dictionaryofkeywordsfrom2.pickle', 'wb') as file:
    pickle.dump(dictionaryofkeywords, file)
```
